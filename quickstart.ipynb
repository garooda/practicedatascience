```
import matplotlib.pyplot as plt
```

```
x = np.array([1,2,3,4,5])
y = (x**2)+3
```

```
print(y)
```

```
plt.scatter(x,y)
plt.xlabel("days")
plt.ylabel("temperatures")
plt.show()
```

```
plt.style.use('seaborn')
```

```
plt.plot(x,y, label="monday")
plt.title("Weather Report")
plt.xlabel("days")
plt.ylabel("temperatures")
plt.legend()
plt.show()
```

Working with OpenCV

```
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```
img = cv2.imread('bulbasaur.png')
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```
print(img.shape)
```

```
plt.imshow(rgb_image)
plt.axis("off")                 #to remove the axis numbers
plt.show()
```

```
rgb_image.shape
```

```
crop_image = rgb_image[50:125, 50:125, :]      #croping the image
```

```
plt.imshow(crop_image)
```

```
crop_image.shape
```

```
new_image = cv2.resize(crop_image, (1000,1000))
```

```
plt.imshow(new_image)
```

```
new_rgb_image = cv2.resize(rgb_image,(448,450))  #resize is used to change the resolution of image
```

```
plt.imshow(new_rgb_image)
plt.axis('off')
plt.show()
```
